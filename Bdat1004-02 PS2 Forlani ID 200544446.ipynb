```python
#Question 1
a = 0
#method to make a global and to call other method c
def b():
    global a
    a = c(a)
#method takes paramater a and adds 2
def c(a):
    return a + 2

#step 1
b()
#step 2
b()
#step 3
b()
#step 4
a
```

As the code above shows, the expression (a) is evaluated as 6. This occurs from the following:
We have set the a value to 0 and created two methods: b() and c(a). 
•	In step 1 we have called on the first method, b() where it will make (a) global and then call on our second method c(a). It is within the method, c(a), that we have the value a + 2 which is 0 + 2 = 2.
•	In step 2, we have done the same thing as step 1 except this time (a )is now 2 and not 0. With this in mind, when we call on the first method b() it will make (a), which is now 2, global. It will then call on the second method c(a). Within this method, c(a), we will return a + 2, which is 2 + 2 = 4. So, now (a) is 4.
•	Step 3 is identical to step 2 and 1 except this time (a) is now 4 and not 0 or 2. With this in mind, when we call on the first method b() it will make (a), which is now 4, global. It will then call on the second method c(a). Within this method, c(a), we will return a + 2, which is 4 + 2 = 6. So, now (a) is 6.
•	In step 4, the final step, we have asked what (a) is. Through steps 1 through 3 we have the value (a) as 6 which is why the output for this code on Jupyter is 6.

```python
#Question 5

def binary(n):
    #if the given value is less than or equal to 1, return the value as a string
    if n <= 1:
        return str(n)
    else:
        #recursives back into the function with n being divided by 2 (no remainder)
        #after it goes through until reaching a value < = 1 it will return and give each
        #time it reached the else with % 2
        return binary(n // 2) + binary(n % 2)

#user inputs a value
input_value = input("Enter a number: ")

#if user inputted an invalid entry - only loop for entire question
while not input_value.isnumeric():
    input_value = input("Entered value is invalid. Please enter a new value: ")

n = int(input_value)
int(binary(n))
```

```python
#Question 6
from html.parser import HTMLParser
#subclass of HTMLParser
class HeadingParser(HTMLParser):
    #constructor for inheritiance and numberofindents to equal 0
    def __init__(self):
        super().__init__()
        self.numberofindents = 0
    
    #in built method which sill look at the starting tag and assign an indent value
    def handle_starttag(self, tag, attrs):
        if tag == "h1":
            self.numberofindents = 1
        elif tag == "h2":
            self.numberofindents = 2
        elif tag == "h3":
            self.numberofindents = 3
        elif tag == "h4":
            self.numberofindents = 4
        elif tag == "h5":
            self.numberofindents = 5
        elif tag == "h6":
            self.numberofindents = 6
        else:
            self.numberofindents = 0
    
    #in built method which will take the data portion in between the start and end tag, and indent based on the assigned
    #value
    def handle_data(self, data):
        #if the indent value assigned from handle_starttag is greater than 0
        if self.numberofindents > 0:
            #empty string
            indentsstring = ""
            #subtracts each assigned indent value from handle_data by 1
            self.numberofindents -= 1
            #Will indent each header by the number of indents (after it has been subtracted by 1 from previous line)
            while (self.numberofindents > 0):
                indentsstring += " "
                #goes through the loop again in case the value is still greater than 0
                self.numberofindents -= 1
            #adds the data to the number of indents
            indentsstring += data
            print(indentsstring)
            

infile = open('w3c.html')
content = infile.read()
infile.close()
hp = HeadingParser()
hp.feed(content)
```

```python
#Question 7
from urllib.request import urlopen
from urllib.parse import urljoin
from urllib.parse import urljoin
from html.parser import HTMLParser

#Creates class which inherits HTMLParser - parses the webpage for links
class Collector(HTMLParser):
    
    #Constructor for collector
    def __init__(self, url):
        HTMLParser.__init__(self)
        self.url = url
        self.links = []
    
    #Method that gets the website links
    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http':
                        self.links.append(absolute)
    
    #Method that returns the website links
    def getLinks(self):
        return self.links

#Method that opens the main URL, reads it and takes the links from that page    
def analyze(url):
    content = urlopen(url).read().decode()
    collector = Collector(url)
    collector.feed(content)
    urls = collector.getLinks()
    
    return urls

#Creates a variable that will be a collection of visited websites
visited = set()

#Method that will print all the webpages visited to a certain assigned depth.
#It will also indent each page based on the depth
def webdir(url, depth, indent):
    global visited
    visited.add(url)
    idx = depth
    i = 0
    url_string = ""
    
    #Adds a certain number of tabs based on the depth value (in order to indent properly)
    while (i < indent):
        url_string += '\t'
        i += 1
    
    #Adds URL to the amount of tabs and prints the string
    url_string += url
    print(url_string)
    
    #calling analyze method to get the rest of the links
    links = analyze(url)

    #for each link 
    for link in links:
        #for links found n depth away
        while (depth > 0):
            if link not in visited:
                #calls on the method webdir to print the unique page
                try:
                    webdir(link, depth - 1, indent + 1)
                #if an error occurs while calling the method
                except:
                    pass
            
            #link has already been printed and will pass to not print duplicates
            else:
                depth = depth - 1
        #resets the depth
        depth = idx

#Goes to link with depth 2 and starting indent of 0
webdir('https://reed.cs.depaul.edu/lperkovic/test1.html', 2, 0)
```

```python
#Question9
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

print("--------------------Part A--------------------")
#empty list which will later be all capitalized words
capitalizedwords = []
#for each word in the list words
for word in words:
    #capitalizes the word
    word_capitalized = word.upper()
    #appends to final list
    capitalizedwords.append(word_capitalized)
print(capitalizedwords)

print("--------------------Part B--------------------")
#empty list which will later be all lower case words
lowerwords = []
#for each word in the list words
for word in words:
    #lower cases the word
    word_lower = word.lower()
    #appends to final list
    lowerwords.append(word_lower)
print(lowerwords)

print("--------------------Part C--------------------")
#empty list which will later be all lengths of the words
lengthwords = []
#for each word in the list words
for word in words:
    length = len(word)
    #appends to final list
    lengthwords.append(length)
print(lengthwords)

print("--------------------Part D--------------------")
#empty list which will later be the final list printed (the list of lists)
finallst = []
#empty list which will later be each words list
eachwordlst = []
#for each word in the list words
for word in words:
    #capitalizes the word
    word_capitalized = word.upper()
    #lower cases the word
    word_lower = word.lower()
    #finds the length of the word
    length = len(word)
    #appends each of the above to each words list
    eachwordlst = [word_capitalized, word_lower, length]
    #appends each list to final list
    finallst.append(eachwordlst)
print(finallst)

print("--------------------Part E--------------------")
#note: question states to print words containing 4 or more characters but the example produces a list that doesn't satisfy
#this requirement - i.e. the words "the", "fox", "dog" are words with 3 characters and not 4.
lstof4 = []
#for each word in the list words
for word in words:
    #if the length of the word is greater than 3 it will append to the list labelled as "lstof4"
    if len(word) > 3:
        #appends to final list
        lstof4.append(word)
print(lstof4)
```
